```python
 #importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets,svm
import numpy as np
import random

 #importing dataset
iris = datasets.load_iris()

X = iris.data[:,:2] #we will be taking only first two rows to make visualizations 2D
y = iris.target # X will be the data to give input and y will be the actual answers for X input
```

```python
print("y (or) targets: \n",y)
print("X (or) Data:\n",X)
```

```python
# Model declaration and fitting : 
#Hint Syntax: model = svm.SVC(kernel = "kernel name",gamma(for rbf) = ,degree(for poly) = , C=C)

C = 1 # regularization factor

# instanciate SVM with kernel rbf with gamma = 0.7 (but you can pick your own gamma)
rbf_svc = svm.SVC(kernel = "rbf", gamma = 0.7, degree = 3, C=C)

# instanciate SVM with kernel poly with degree = 3 (but you can pick your own degree)
poly_svc =  svm.SVC(kernel = "poly", gamma = 0.7, degree = 3, C=C)


# Model fitting
rbf_svc.fit(X,y)
poly_svc.fit(X,y)

# storing the models for the function below
models = [rbf_svc, poly_svc]
```

```python
def plot_graph(clf): # clf will be the model here (the input must be either rbf_svc or poly_svc(dont change the spelling))
    
    h = 0.2
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
    
    titles = ['SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel']

    if clf in models:
        k = models.index(clf)
    
    else:
        print("check again")
        return
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    color = plt.cm.Set1
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=color, alpha=0.6)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=color , s = 30)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[k])
    plt.show()
```

```python
plot_graph(poly_svc) # shows graph for svm classification of svm with poly kernel
```

```python
plot_graph(rbf_svc)  # shows graph for svm classification of svm with rbf kernel
```
